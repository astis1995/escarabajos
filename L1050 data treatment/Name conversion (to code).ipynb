```python
### This code converts transflectance data to a general file name
### And adds TXRX
import pandas as pd
import numpy as np
import matplotlib
import os

#code name samples path
code_name_path = r"C:\Users\EstebanSoto\Documents\Estudio Optico Escarabajos\CODE NAME SAMPLES.txt"
#Lists all archives in folder
transflectance_folder_data_path = r"C:\Users\EstebanSoto\Documents\Estudio Optico Escarabajos\new_data\2023NOV_TRANSFLECTANCE\transflectance"
file_list = os.listdir(transflectance_folder_data_path)
file_list
```

```python
# filters a list of strings to create a new list containing only the elements that end with ".txt"

def filter_asc_elements(input_list):
    return [element for element in input_list if element.endswith(".ASC")]
def filter_substring_elements(path_strings, substring):
    filtered_paths = [path for path in path_strings if substring in path]
    return filtered_paths
# Filtering elements ending with ".ASC"
result_list = filter_asc_elements(file_list)

# Displaying the filtered list
print(result_list)
print(len(result_list))
```

```python
#read codes
codes_df = pd.read_csv(code_name_path, sep="	",header=0,  
                     names=["code", "description"]) 
codes_df
```

```python
def get_code_from_description(description, dataframe):
    code = dataframe.loc[dataframe['description'] == description, 'code'].values
    if len(code) > 0:
        return code[0]
    else:
        return None
```

```python
for description in codes_df['description']:
    temp_list = filter_substring_elements(result_list, description)
    #print(description)
    #print(temp_list)
    for filename in temp_list:
        print(filename)
        
        basename = os.path.basename(filename).replace(".ASC", "").replace("TXRX", "")
        print(basename)
        string = basename
        
        number = string.split('-', 1)[0] + '-' if string[0].isdigit() else '' 
        print(number)
        print(len(number))
        
        new_filename = filename.replace(description, get_code_from_description(description, codes_df) ).strip()
        if(len(number)==0):
            new_filename = "1-" + new_filename
        print(new_filename)
        
        #print(new_filename)
        #filename = transflectance_folder_data_path +"\\"+file
        #print(filename)
        df = pd.read_csv(transflectance_folder_data_path + "\\"+ filename, sep="	",header=90,  
                         names=["wavelength", "transflectance"])
        #print(df)

        corrected_path = transflectance_folder_data_path +"\\" + "new name\\"

        if not os.path.exists(corrected_path):
            os.mkdir(corrected_path)

        #saves file
        #print(get_code_from_description(file, codes_df))
        
        my_numpy = df.to_numpy()
        new_archive_name = corrected_path + new_filename
        np.savetxt(new_archive_name, my_numpy,fmt='%s', delimiter="\t")
        #print(temp_list)
    
```
