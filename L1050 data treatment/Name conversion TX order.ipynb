```python
### This code converts transflectance data to a general file name
### And adds TXRX
import pandas as pd
import numpy as np
import matplotlib
import os

#code name samples path
code_name_path = r"E:\CICIMA\Estudio Optico Escarabajos\CODE NAME SAMPLES.txt"
#Lists all archives in folder
transflectance_folder_data_path = r"E:\CICIMA\2023_Q4_TRANSMITTANCE_AND_TRANSFLECTANCE\2023_Q4_TRANSMITTANCE_AND_TRANSFLECTANCE\2023NOV_TRANSFLECTANCE\ASC"
file_list = os.listdir(transflectance_folder_data_path)
#file_list
```

```python
# filters a list of strings to create a new list containing only the elements that end with ".txt"

def filter_asc_elements(input_list):
    return [element for element in input_list if element.endswith(".ASC")]
def filter_substring_elements(path_strings, substring):
    filtered_paths = [path for path in path_strings if substring in path]
    return filtered_paths
# Filtering elements ending with ".ASC"
result_list = filter_substring_elements(file_list, ".ASC")

# Displaying the filtered list
print(result_list)
```

```python
#read codes
codes_df = pd.read_csv(code_name_path, sep="	",header=0,  
                     names=["code", "description"]) 
codes_df
```

```python
def get_code_from_description(description, dataframe):
    code = dataframe.loc[dataframe['description'] == description, 'code'].values
    if len(code) > 0:
        return code[0]
    else:
        return None
```

```python
def get_file_number(file_str):
        import re
        #print("string")
        #print(file_str)
        re1 = r"\w*(\d)+-*\w+(.ASC)*(.txt)*"
        #Names are in the form CODE-MEASUREMENTNUMBER.TXT
        p = re.compile(re1)
        m = p.match(file_str)
        #print("match")
        #print(m)
        if m:
            #print(f"return {m.group(1)}")
            return(m.group(1))

        return ""
```

```python
get_file_number("TXRX1-C RESPLENDENS.ASC")
```

```python
def rearrange_filename(filename):
        filename = filename.replace("..", ".").replace(".ASC", "")
        #print(filename)
        parts = filename.split('-')
        for idx, part in enumerate(parts):
            #print(part)
            if part.startswith('TX'):
                prefix = part
                #print(f"2prefix {prefix}")
                suffix = parts[idx-1]
                prefix = get_file_number(filename)
                #print(f"2suffix {suffix}")
                rearranged_filename = f"{suffix}-{prefix}.ASC"
                return rearranged_filename

# Example filenames
filenames = ["C.AURI.INBI-TX1.ASC", "C.AURI.INBI-TX3.ASC", "C.AURI.INBI-TX2.ASC"]

# Rearrange filenames
rearranged = rearrange_filename("C.AURI.INBI-TX1.ASC")
print(rearranged)
```
