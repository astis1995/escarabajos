```python
#pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Datos de ejemplo
longitudes_de_onda = np.linspace(380, 750, 100)  # Longitudes de onda en nm
intensidad = np.random.rand(100)  # Intensidad aleatoria

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Crear un fondo de arcoris
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

# Define a custom colormap from violet to red
colors = ["#8A2BE2", "#FF0000"]  # Violet and red
custom_cmap = LinearSegmentedColormap.from_list("violet_red", colors, N=256)

# Mostrar el fondo
ax.imshow(gradient, aspect='auto', cmap=custom_cmap, extent=[400, 700, 0, 1], alpha=0.3)

# Graficar los datos
ax.plot(longitudes_de_onda, intensidad, color='black')

# Etiquetas y ttulo
ax.set_xlabel('Longitud de Onda (nm)')
ax.set_ylabel('Intensidad')
ax.set_title('Grfica de Longitudes de Onda con Fondo de Arcoris')

plt.show()
```
