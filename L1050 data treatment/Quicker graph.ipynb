```python
# Quicker graph

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
import math

#This line of code allow us to access data in colab
#from google.colab import drive
#drive.mount("/content/drive")
```

```python
#base folder
base_folder = r"C:\Users\EstebanSoto"

#Image location
save_location = os.path.join(base_folder, "quick_graph_images")

#Location of the file that you want to plot
file_location = r"C:\Users\EstebanSoto\Downloads\HapLIR\HapL_ALE\ALE puro_1.csv"
```

```python
def create_path_if_not_exists(path):
        # Check if the path already exists
        if not os.path.exists(path):
            # Create the directory and any missing parent directories
            os.makedirs(path)
            print(f"Directory '{path}' created successfully.")
        else:
            print(f"Directory '{path}' already exists.")
```

```python
df = pd.read_csv(file_location, header = 1, names=["wn","T"])
```

```python
x = df["wn"]
y = df["T"]

plt.figure(figsize=(15,6))
#plt.grid()
colorg = "green"
plt.plot(x,y, color = colorg, linewidth = 1)
plt.title("HAp-L-Alendronate IR Spectrum")
plt.xlabel("wavenumber [cm$^{-1}$")
plt.ylabel("transmittance [%]" )
filename = "HPL.svg"
create_path_if_not_exists(save_location)
plt.savefig(os.path.join(save_location,filename))
plt.show()
```
