```python
import pandas as pd

class Specimen_Collection:
    def __init__(self, name, path):
        self.name = name
        self.path = path
        self.df = pd.read_csv(path, sep="\t", decimal =".", header=0, encoding="iso-8859-1") # Assuming the file is in CSV format
        self.df.loc[:, ["collection"]] = self.name
    def display_data(self):
        print(self.df)
        
    def search_specimen(self, column, value):
        result = self.df[self.df[column] == value]
        if len(result) > 0:
            print("Specimen found:")
            print(result)
        else:
            print("Specimen not found.")
```

```python
collections = [Specimen_Collection("CICIMAUCR",cicimaucr_database).df , Specimen_Collection("INBUCR",inbucr_database).df, Specimen_Collection("ANGSOL", angsol_database).df, Specimen_Collection("BIOUCR", bioucr_database).df]
```

```python

unified_collection = pd.concat(collections, ignore_index = True)
type(unified_collection)
for column in unified_collection.columns: 
    print(type(unified_collection.loc[0, :]       ))
```

```python
#save unified
data = unified_collection.to_numpy()
header = "\t".join(unified_collection.columns)
print(f"header: {header}")
archive_name = save_location + "\\" + database_descriptor +".txt"
print(archive_name)
with open(archive_name, "wb") as f: 
    np.savetxt(archive_name, data, fmt = "%s", delimiter = "\t", comments = "", header = header, encoding="latin1")
```
