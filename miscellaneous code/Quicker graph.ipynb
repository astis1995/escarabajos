```python
#base folder
base_folder = r"E:\Estudio-espectral-escarabajos\PruebaLinealidadrespuesta"

#Image location
save_location = os.path.join(base_folder, "quick_graph_images")

#Location of the file that you want to plot
file_location = r"E:\Estudio-espectral-escarabajos\PruebaLinealidadrespuesta\LCP.csv"
file_location2 = r"E:\Estudio-espectral-escarabajos\PruebaLinealidadrespuesta\RCP.csv"
```

```python
def create_path_if_not_exists(path):
        # Check if the path already exists
        if not os.path.exists(path):
            # Create the directory and any missing parent directories
            os.makedirs(path)
            print(f"Directory '{path}' created successfully.")
        else:
            print(f"Directory '{path}' already exists.")

def list_files_in_folder(folder_path):
    try:
        # Get list of files and directories
        files = os.listdir(folder_path)

        # Filter out directories, show only files
        files = [f for f in files if os.path.isfile(os.path.join(folder_path, f))]

        return files
    except FileNotFoundError:
        return f"Error: The folder '{folder_path}' does not exist."
    except PermissionError:
        return f"Error: Permission denied to access '{folder_path}'."
```

```python
df = pd.read_csv(file_location, header = 9, sep= ",", decimal = ".", names=["wn","T"])
df2 = pd.read_csv(file_location2, header = 9, sep= ",", decimal = ".", names=["wn","T"])
```

```python
df = pd.read_csv(file_location, header = 9, sep= ",", decimal = ".", names=["wn","T"])
df2 = pd.read_csv(file_location2, header = 9, sep= ",", decimal = ".", names=["wn","T"])
x = df["wn"]
y = df["T"]
x2 = df2["wn"]
y2 = df2["T"]

plt.figure(figsize=(15,6))
#plt.grid()
colorg = "green"
plt.plot(x,y, color = colorg, linewidth = 1)
plt.plot(x2,y2, color = "r", linewidth = 1)
plt.title("HAp-L-Alendronate IR Spectrum")
plt.xlabel("wavenumber [cm$^{-1}$")
plt.ylabel("transmittance [%]" )
filename = "HPL.svg"
create_path_if_not_exists(save_location)
plt.savefig(os.path.join(save_location,filename))
plt.show()
```
