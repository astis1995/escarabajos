```python
def list_files_in_folder(folder_path):
    files_only_in_folder = []
    
    # Get the list of all files in the folder and its subfolders
    all_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            all_files.append(os.path.join(root, file))
    
    # Get the list of files that are only in the top-level folder
    for file in all_files:
        if os.path.isfile(file) and os.path.dirname(file) == folder_path:
            files_only_in_folder.append(file)
    
    return files_only_in_folder
```

```python
def read_dataframes(file_path_list, measurement_type): 
    
    dataframes = []
    
    for file_path in file_path_list: 
        
        df = pd.read_csv(file_path, sep='\t', header=1, names=['wavelength', measurement_type]).dropna()

        #Replace commas with dots
        df = df.replace(',','.', regex=True)
        
        #print(df)
        df["file"] = file_path
        df["sample"] = os.path.basename(file_path)
        df["measurement_type"] = measurement_type
        
        #convert columns to float
        df['wavelength'] = df['wavelength'].astype(float)
        df[measurement_type] = df[measurement_type].astype(float)
        
        # Append the dataframe to the list
        dataframes.append(df)
        
    return dataframes 
```

```python
def correlation_matrix(df1, df2):
   
    # Calculate the correlation matrix
    correlation_matrix = df1.corrwith(df2, axis=0)
    
    return correlation_matrix
```

```python
def get_numeric_cols(df):
    # Select only numeric columns
    numeric_columns = df.select_dtypes(include='number')

    # Create a new DataFrame with only numeric columns
    numeric_df = df[numeric_columns.columns]
    
    return numeric_df
```

```python
def copy_metadata(df1, df2, df_result):
    df_result["file1"] = df1["file"].iloc[0]
    df_result["file2"] = df2["file"].iloc[0]
    df_result["sample1"] = df1["sample"].iloc[0]
    df_result["sample2"] = df2["sample"].iloc[0]
    df_result["measurement_type_1"] = df1["measurement_type"].iloc[0]
    df_result["measurement_type_2"] = df2["measurement_type"].iloc[0]
    return df_result
```

```python
def correlations(dataframe_list_1, dataframe_list_2):
    corrs = []
    #print(len(dataframe_list_1))
    combination_list = []
    
    for dataframe1 in dataframe_list_1:
        
        for dataframe2 in dataframe_list_2:
            correlation_df = correlation_matrix(get_numeric_cols(dataframe1) , get_numeric_cols(dataframe2))
            correlation_df_with_metadata = copy_metadata(dataframe1, dataframe2, correlation_df)
            #file	sample	measurement_type
            corrs.append(correlation_df_with_metadata)

    corrs_df = pd.DataFrame(corrs)
    return corrs_df
```

```python
def top_correlations(df, number,measurement_type):
    # Assuming 'df' is your DataFrame
    # Filter the DataFrame where sample1 and sample2 are different
    filtered_df = df[df['sample1'] != df['sample2']]

    # Sort the filtered DataFrame by transmittance column in descending order and get the top 10 rows
    top_transmittance = filtered_df.sort_values(by=measurement_type, ascending=False).head(number).reset_index(drop=True)
    #filtered_rows = pd.DataFrame(columns=top_transmittance.columns)
    #filtered_rows= select_unique_rows(filtered_rows, "transmittance")
    filtered_df = top_transmittance[ top_transmittance.index % 2 != 0 ]
    return (filtered_df)
```

```python
def filter_dataframes_by_range(min_wavelength, max_wavelength, dataframe_list):
    filtered_dataframes = []
    
    for df in dataframe_list:
        # Assuming the wavelength column in each DataFrame is named 'wavelength'
        filtered_df = df[(df['wavelength'] >= min_wavelength) & (df['wavelength'] <= max_wavelength)]
        filtered_dataframes.append(filtered_df)
    
    return filtered_dataframes
```
