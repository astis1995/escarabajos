```python
#get_rolo_coefficients
rolo_coefficients_df = pd.read_csv(rolo_coefficients_path, sep='\t', header=0, names=["wavelength",'a0', "a1", "a2", "a3", "b1", "b2", "b3", "d1", "d2", "d3"]).dropna()
```

```python
#constants
markersize = 2
p1 = 4.06054
p2 = 12.8802
p3 = -30.5858
p4 = 16.7498
c1 = 0.00034115
c2=-0.0013425
c3 = 0.00095906
c4 = 0.00066229
OmegaM = 6.4177 * np.power(10.0, -5.0)
```

```python
#Read sun irradiance
sun_irradiance_df = pd.read_csv(sun_irradiance_path, sep='\t', header=0, names=['wavelength', "irradiance"]).dropna()
```

```python
sun_irradiance_df
```

```python
#get the whole number part of wavelength
sun_irradiance_df["wavelength_whole_part"] = np.floor(sun_irradiance_df["wavelength"].to_numpy()).astype(int)
sun_irradiance_df = sun_irradiance_df.groupby(by="wavelength_whole_part").mean().sort_values(by=["wavelength"], ascending=True)
sun_irradiance_df
```

```python
def get_coefficients(wavelength):
    # Sort the DataFrame by 'wavelength' in descending order
    sorted_df = rolo_coefficients_df[rolo_coefficients_df["wavelength"] <= wavelength].sort_values(by="wavelength", ascending=False)
    #print(rolo_coefficients_df)
    #print(sorted_df)
    
    if len(sorted_df) == 0 or sorted_df.empty: 
        return rolo_coefficients_df[0]
    # Get the first row from the sorted DataFrame
    closest_row = sorted_df.iloc[0]  
    
    return closest_row
```

```python
get_coefficients(500)["a0"]
```

```python
def Ak_calculation(g, theta, phi, big_phi, wavelength):
    coeff = get_coefficients(wavelength)
    aterms = coeff["a0"] +coeff["a1"]*np.power(g, 1.0)+coeff["a2"]*np.power(g, 2.0)+coeff["a3"]*np.power(g, 3.0)
    bterms = coeff["b1"]*np.power(big_phi, 1.0) +coeff["b2"]*np.power(big_phi, 3.0)+coeff["b3"]*np.power(big_phi, 5.0)
    cterms = c1*theta +c2*phi + c3*theta*big_phi +c4*big_phi*phi
    dterms = coeff["d1"]*np.exp(-g/p1)+  coeff["d2"]*np.exp(-g/p2)+ coeff["d3"]*np.cos((g-p3)/p4)
    
    lnAk = aterms+bterms+cterms+dterms
    Ak = np.exp(lnAk)
    #print(lnAk)
    return Ak
```

```python
type(get_coefficients(351)["a0"])
```

```python
g = 90.0* np.pi /180.0 #Absolute phase angle (radians) 0 is full moon, 180 is new moon
theta = 0.0* np.pi/180.0 #selenographic latitude of the observer (radians)
phi = 0.0* np.pi/180.0 #selenographic longitude of the observer (radians)
big_phi = 0.0* np.pi/180.0 #selenographic longitude of the sun (radians)
wavelength= 1000.0 
print(Ak_calculation(g, theta, phi, big_phi, wavelength))
```

```python
wavelengths = np.linspace(350,997,997-350)
wavelengths
```

```python
Ak_list = []
for wv in wavelengths: 
    #print(wv)
    Ak = Ak_calculation(g, theta, phi, big_phi, wv)
    Ak_list.append(Ak)
ak_df = pd.DataFrame([])
ak_df["wavelength"] = wavelengths
ak_df["ak"] = Ak_list
ak_df
```

```python
#We'll change the index of each dataframe to its wavelength
ak_df["wavelength_whole_part"] = np.floor(ak_df["wavelength"].to_numpy()).astype(int)
ak_df.set_index('wavelength_whole_part', inplace=True)
sun_irradiance_df["wavelength_whole_part"] =np.floor(sun_irradiance_df["wavelength"].to_numpy()).astype(int)
sun_irradiance_df.set_index('wavelength_whole_part', inplace=True)
```

```python
sun_irradiance_df
```

```python
#Now we multiply the ak by OmegaM, the irradiance of the sun and divide it by pi
ik_df = pd.DataFrame([])
ak_df["ak"]
sun_irradiance_df["irradiance"] 

# Multiply columns from different DataFrames based on matching indices
result = ak_df.loc[:, 'ak'].mul(sun_irradiance_df.loc[:, 'irradiance']) *OmegaM / np.pi

# Create a new DataFrame with the result
moon_spectral_irradiance=pd.DataFrame([])
moon_spectral_irradiance["wavelength"]=sun_irradiance_df["wavelength"]
moon_spectral_irradiance["irradiance"] = result
moon_spectral_irradiance = moon_spectral_irradiance.dropna()
print(moon_spectral_irradiance)
```
