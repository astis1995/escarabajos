```python

cupy_available = False
try:
    import cupy as cp  # Import CuPy for GPU computations
    cupy_available = True
except:
    import numpy as cp
    print("No cupy available")
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from tqdm import tqdm  # Import tqdm for progress bar
import time
import numpy as np
from scipy.special import logsumexp

initial_time = time.time()

# Parameters
L = 0.25  # 10 cm side
Lz = 0.02
Nx = Ny = Nz = 80  # 1mm resolution
dx = L / (Nx - 1)
dy = L / (Ny - 1)
dz = Lz / (Nz - 1)
total_secs = 0.1  # s
dt = 0.001
sigma = 0.01 #gaussians

# Constants for the heat transfer equation
rho = 100   # Density, kg/m^3
c = 1000    # Specific heat capacity, J/(kg*K)
epsilon = 0.9  # Emissivity
SBc = 5.67e-8  # Stefan-Boltzmann constant
g_amplitude = (40)  # Laser temperature when max, K
T0 = (273)  # Ambient temperature, K
h = 10      # Convective heat transfer coefficient
k = 0.5     # Thermal conductivity

num_snapshots = 60  # You can set the number of snapshots here

# Initial condition: Gaussian distribution
def create_gaussian_distribution(Nx, Ny, Nz, amplitude, sigma, T0):
    center = (Nx // 2, Ny // 2, Nz // 2)
    x = cp.linspace(0, L, Nx)
    y = cp.linspace(0, L, Ny)
    z = cp.linspace(0, L, Nz)
    X, Y, Z = cp.meshgrid(x, y, z, indexing='ij')
    gaussian = amplitude * cp.exp(-(((X - x[center[0]])**2 + (Y - y[center[1]])**2 + (Z - z[center[2]])**2) / (2 * sigma**2)))
    return gaussian + T0

# Laplacian function
def laplacian_3d(u):
    return ((cp.roll(u, -1, axis=0) - 2 * u + cp.roll(u, 1, axis=0)) / dx**2 +
            (cp.roll(u, -1, axis=1) - 2 * u + cp.roll(u, 1, axis=1)) / dy**2 +
            (cp.roll(u, -1, axis=2) - 2 * u + cp.roll(u, 1, axis=2)) / dz**2)

# Heat transfer equation


def stable_radiation(u, T0, epsilon):
    # Calculate 4 * log(u) for the entire array and flatten to 1D
    u_log = 4 * np.log(u).flatten().get()
    #print(len(u_log))
    # Create a constant array for 4 * log(T0) and flatten to 1D, multiply for -1
    T0_log = -np.full_like(u_log, 4 * np.log(T0), dtype=np.float64)
    #print(len(T0_log))
   
    # Use logsumexp to calculate exp(u_log) - exp(T0_log) in a stable way
    radiation_term_flat = np.exp(logsumexp([u_log, T0_log], axis = 0))
   
    # Reshape back to the original shape
    radiation_term = radiation_term_flat.reshape(u.shape)
   
    # Multiply by the constants epsilon and sigma
    return epsilon * SBc * radiation_term

   
def heat_transfer_equation(u, epsilon):
    radiation = cp.array(stable_radiation(u, T0, epsilon))
    convection = h * (u - T0)
    conduction = k * laplacian_3d(u)
    ret = (-(1 / (rho * c)) * (radiation + convection + conduction))
    return ret

# RK4 Step function with boundary condition enforcement
def rk4_step_3d(u, dt, epsilon, equation_func, T_const):
    k1 = equation_func(u, epsilon)
    k2 = equation_func(u + 0.5 * dt * k1, epsilon)
    k3 = equation_func(u + 0.5 * dt * k2, epsilon)
    k4 = equation_func(u + dt * k3, epsilon)
    u_new = u + (dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)

    # Enforce boundary conditions
    u_new[0, :, :] = T_const    # x = 0 boundary
    u_new[-1, :, :] = T_const   # x = L boundary
    u_new[:, 0, :] = T_const    # y = 0 boundary
    u_new[:, -1, :] = T_const   # y = L boundary
    u_new[:, :, 0] = T_const    # z = 0 boundary
    u_new[:, :, -1] = T_const   # z = L boundary
    return u_new


# Calculate maximum steps for a 10-second limit
max_steps = 60
print(f"Estimated maximum number of steps for a 10-second run: {max_steps}")

# Set up snapshots
snapshot_indices = cp.linspace(0, max_steps - 1, num_snapshots, dtype=int)

# Reset initial condition and snapshots
u = create_gaussian_distribution(Nx, Ny, Nz, g_amplitude, sigma, T0).astype(cp.float64)
snapshots = []

# Run simulation and collect snapshots
print("Running simulation:")
for t in tqdm(range(max_steps), desc="Simulation Progress"):
    u = rk4_step_3d(u, dt, epsilon, heat_transfer_equation, T0)
    u 
    if t in snapshot_indices:
        #print(type(u))
        if isinstance(u, np.ndarray):
            snapshots.append(u)  # Move data from GPU to CPU for storage
        else:
            snapshots.append(u.get())
           
# Plotting snapshots and creating an animation
fig, ax = plt.subplots(figsize=(8, 8))

if cupy_available:
    x = cp.asnumpy(cp.linspace(0, L, Nx))
    y = cp.asnumpy(cp.linspace(0, L, Ny))
else:
    x = cp.linspace(0, L, Nx)
    y = cp.linspace(0, L, Ny)
# Create the meshgrid
X, Y = np.meshgrid(x, y)

# Initial contour plot
z_slice = snapshots[0][:, :, Nz // 2]  # Middle slice in z-direction
contour = ax.contourf(X, Y, z_slice, levels=20, cmap='hot')
cbar = plt.colorbar(contour)  # Associate colorbar with the contour plot

# Update function for the animation
# Update function for the animation
def update(frame):
    ax.clear()  # Clear the axes to update the plot
    
    # Get the middle slice in the z-direction for the current frame
    z_slice = snapshots[frame][:, :, Nz // 2]
    
    # Determine the min and max of the current slice to adjust the contour levels dynamically
    vmin = np.min(z_slice)
    vmax = np.max(z_slice)
    
    # Create a new contour plot with dynamically adjusted levels
    contour = ax.contourf(X, Y, z_slice, levels=20, cmap='hot', vmin=vmin, vmax=vmax)
    
    # Update title and labels
    ax.set_title(f'Time (ms): {snapshot_indices[frame] * dt * 1000}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    
    # Optional: Focus on the inner quadrant by setting limits (uncomment to use)
    # ax.set_xlim((L / 4, 3 * L / 4))  # Half of the total size to the full size
    # ax.set_ylim((L / 4, 3 * L / 4))  # Half of the total size to the full size
    
    # Update the colorbar to match the new contour plot
    cbar.update_normal(contour)


# Set up the animation
ani = animation.FuncAnimation(fig, update, frames=len(snapshots), repeat=False)

# Save the animation as a GIF
ani.save('simulation_video.gif', writer='pillow', fps=3)

# Show the plot
plt.show()

final_time = time.time()
print(f"It took {final_time - initial_time=}")
```
