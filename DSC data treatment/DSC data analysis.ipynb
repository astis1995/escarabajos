```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
base_folder = r"C:\Users\EstebanSoto\Jupyter\escarabajos\DSC_data\V. Soto"
filename1 = "DSC_A1_1_0-100C_6Cmin_2.txt"
filename2 = "DSC_A1_1_0-100C_6Cmin_TXT.txt"
filename3 = "DSC_A1_2_0-100C_6Cmin_TXT.txt"
file_path = os.path.join(base_folder, filename1)

# #filename2 = 
# df = pd.read_csv(file_path, sep="\t", header = 9 , names = ["time", "T", "phi"], decimal=",")
# df = df.dropna()
# print(df)
# columns = df.columns
# for col in columns:
#     df[col] = df[col].replace(',', '.', regex=True)
# print(df)


# Assuming you know the first and last line numbers of the first table
first_line = 11  # Replace with the actual line number
last_line = 10012   # Replace with the actual line number

# Read the lines of interest from the file
lines = []
with open(file_path, 'r') as file:
    for line_num, line in enumerate(file, start=1):
        if line_num >= first_line and line_num <= last_line:
            lines.append(line.strip())

# Create a DataFrame from the lines
df = pd.DataFrame([line.split('\t') for line in lines])

# Optionally, specify column names if available
df.columns = ["time", "T", "phi"]
columns = df.columns
for col in columns:
     df[col] = df[col].replace(',', '.', regex=True)
print(df)

df = df.dropna()
df = df.astype(float)
```

```python
time = df["time"]
phi = df["phi"]
plt.figure()
plt.plot(time,phi)
plt.fill_between(time, phi, color='skyblue', alpha=0.4, label='Area under f(x)')
plt.xlabel("time")
plt.ylabel("heat flux")
plt.grid()
plt.show()
```
